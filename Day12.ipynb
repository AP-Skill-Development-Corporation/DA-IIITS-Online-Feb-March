```python
li=[3,4,5,6,20,25,30,13,16]
plt.plot(li) # plotting the values
plt.show()
```

```python
#try with multiple list of values
f=[4,6,8,9,10]
s=(10,12,15,16,18)
plt.plot(f,s) # representing(x,y)
plt.show()
```

```python
#try with multiple list of values
f=[4,6,8,9,10]
s=(10,12,15,16,18)
th=[4,12,34,10,25]
plt.plot(f,s,label='first') #
plt.plot(f,th,label="second")
plt.plot(s,th,label='third')
plt.legend() # location of label/default at loc=1 (top right of the plane)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Multiple Graphs in a plane")
plt.show()
```

```python
# pie plot
x=[50,25,30,24]
plt.pie(x)
plt.show()
```

```python
#adding labels
fdb=['average','good','above average','excellent']
plt.pie(x,labels=fdb,colors=['green','cyan','blue','pink'],
       autopct="%.f%%")
plt.title("Sort by Responses")
plt.show()
```

```python
#adding labels
fdb=['average','good','above average','excellent']
ex=[0,0,0,1]
plt.pie(x,labels=fdb,colors=['green','cyan','blue','pink'],
       autopct="%.f%%",explode=ex)
plt.title("Sort by Responses")
plt.show()
```

```python
#stackplot
#one by one 
i=[9,3,5,4,8]
j=[9,4,5,6,2]
k=[6,8,2,3,1]
c=np.array([8,5,6,4,10])
plt.stackplot(i,k,j,c)
plt.show()
```
